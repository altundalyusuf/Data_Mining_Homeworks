```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#https://www.google.com/covid19/mobility/
url='https://drive.google.com/file/d/18gyHbx6rfogq3yQ-GR9COjcGgyYlCnBZ/view?usp=sharing'
url2='https://drive.google.com/uc?id=' + url.split('/')[-2]
df = pd.read_csv(url2)
df.info()
```

```python
#Soru 2 Cevap:
fig, axs = plt.subplots(5,3)
liste=["retail_and_recreation_percent_change_from_baseline", "grocery_and_pharmacy_percent_change_from_baseline","parks_percent_change_from_baseline", "transit_stations_percent_change_from_baseline","workplaces_percent_change_from_baseline","residential_percent_change_from_baseline"]
i=0
j=0
k=1
y=0
for x in range(0,5,1):
    t=k
    for y in range(k,6,1):
        if(j<3  and i<5 ):
            axs[i,j].scatter(df[liste[x]],df[liste[y]])
            j=j+1 
    
        elif y>5:
            break
        else: 
            j=0
            i=i+1
            axs[i,j].scatter(df[liste[x]],df[liste[y]])
            j=j+1
                 
    k=t+1 
    
for p in range(0,5,1):
    for n in range(0,3,1):
        if(axs[p,n] == axs[1,1] or axs[p,n] == axs[2,2] or axs[p,n] == axs[3,2] or axs[p,n] == axs[4,1] or axs[p,n] == axs[4,2] ):
            axs[p,n].set_xlabel('Negative')
        else:
            axs[p,n].set_xlabel('Positive')  
            
plt.tight_layout()

    
```
